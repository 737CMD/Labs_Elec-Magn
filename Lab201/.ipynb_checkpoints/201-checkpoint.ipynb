```python
x = np.linspace(70, 160, 100)
fig, axs = plt.subplots(1,1, figsize = (11,11))
axs.set_xlim(70, 160)
axs.set_ylim(0, 38)
axs.set_title("Анодная статическая характеристика Ia = f(Ua)")
axs.set_xlabel("Ua, В")
axs.set_ylabel("Ia, mА")
axs.minorticks_on()
kanodstat = np.array([])
for i in range(0, 18):
    axs.scatter(Ua, statIa[i], s = 50, color = "black", marker = "+")
    k = np.polyfit(Ua, statIa[i], 1)
    y = x*k[0] + k[1]
    kanodstat = np.append(kanodstat, k[0])
    axs.plot(x, y, label = f"При Uc = {Us[i]}")
axs.grid(which = 'both', linestyle='--', linewidth = 0.5)
axs.legend()
plt.show()
```

```python
x = np.linspace(70, 160, 100)
fig, axs = plt.subplots(1,1, figsize = (10,10))
axs.set_xlim(70, 160)
axs.set_ylim(0, 14)
axs.set_title("Анодная динамическая характеристика Ia = f(Ua)")
axs.set_xlabel("Ua, В")
axs.set_ylabel("Ia, mА")
axs.minorticks_on()
kanoddynam = np.array([])
for i in range(0, 18):
    axs.scatter(Ua, dynamIa[i], s = 50, color = "black", marker = "+")
    k = np.polyfit(Ua, dynamIa[i], 1)
    kanoddynam = np.append(kanoddynam, k[0])
    y = x*k[0] + k[1]
    axs.plot(x, y, label = f"При Uc = {Us[i]}")
axs.grid(which = 'both', linestyle='--', linewidth=0.5)
axs.legend()
plt.show()
```

```python
statIaset = statIa.transpose()
dynamIaset = dynamIa.transpose()
```

```python
fig, axs = plt.subplots(1,1, figsize = (10,10))
axs.set_xlim(-8, 1.5)
axs.set_ylim(0, 38)
axs.set_title("Сеточная статическая характеристика Ia = f(Uс)")
axs.set_xlabel("Uс, В")
axs.set_ylabel("Ia, mА")
axs.minorticks_on()
ksetstat = np.array([])
for i in range(0, 8):
    axs.scatter(Us, statIaset[i], s = 30, color = "black", marker = "+")
    k = np.polyfit(Us[2+i//4:], statIaset[i][2+i//4:], 3)
    x = np.linspace(-8, 1-(2+i//4)*0.5, 100)
    y = x**3*k[0]+x**2*k[1] + x**1*k[2] + k[3]
    k1 = np.polyfit(Us[:3+i//4], statIaset[i][:3+i//4], 1)
    ksetstat = np.append(ksetstat, k1[0])
    x1 = np.linspace(1-(2+i//4)*0.5, 1, 100)
    y1 = k1[0]*x1+k1[1]
    axs.plot(np.append(x, x1), np.append(y, y1), label = f"При Ua = {Ua[i]}")
axs.grid(which = 'both', linestyle='--', linewidth=0.5)
axs.legend()
plt.show()
print(ksetstat)
```

```python
x = np.linspace(-8,-1, 100)
fig, axs = plt.subplots(1,1, figsize = (10,10))
axs.set_xlim(-8, 1.5)
axs.set_ylim(0, 15)
axs.set_title("Сеточная динамическая характеристика Ia = f(Uс)")
axs.set_xlabel("Uс, В")
axs.set_ylabel("Ia, mА")
axs.minorticks_on()
ksetdynam = np.array([])
for i in range(0, 8):
    axs.scatter(Us, dynamIaset[i], s = 30, color = "black", marker = "+")
    k = np.polyfit(Us[3+i//4:], dynamIaset[i][3+i//4:], 3)
    x = np.linspace(-8, 1-(3+i//4)*0.5, 100)
    y = x**3*k[0]+x**2*k[1] + x**1*k[2] + k[3]
    k1 = np.polyfit(Us[:3+i//4], dynamIaset[i][:3+i//4], 1)
    x1 = np.linspace(1-(2+i//4)*0.5, 1, 100)
    y1 = k1[0]*x1+k1[1]
    ksetdynam = np.append(ksetdynam, k1[0])
    axs.plot(np.append(x, x1), np.append(y, y1), label = f"При Ea = {Ua[i]}")
axs.grid(which = 'both', linestyle='--', linewidth=0.5)
axs.legend()
plt.show()
```

```python
row1 = pd.Series(1/kanodstat[:5], index=Us[:5])
d = {"Стат. Ri": row1}
statRi = np.mean(1/kanodstat[:5])
statSigmaRi = np.var(1/kanodstat[:5])
print(f'Стат. <Ri> = {round(np.mean(1/kanodstat[:5]), 2)}±{round(np.var(1/kanodstat[:5]), 2)} кОм')
pd.DataFrame.from_dict(d, orient="index")
```

```python
row1 = pd.Series(ksetstat, index=Ua)
d = {"Стат. S": row1}
statSi = np.mean(ksetstat)
statSigmaSi = np.var(ksetstat)
print(f'Стат. <Si> = {round(np.mean(ksetstat), 2)}±{round(np.var(ksetstat), 2)} мА/В')
pd.DataFrame.from_dict(d, orient="index")
```

```python
row1 = pd.Series(1/kanoddynam[:5], index=Us[:5])
d = {"Динам. Ri": row1}
print(f'Динам. <Ri> = {round(np.mean(1/kanoddynam[:5]), 2)}±{round(np.var(1/kanoddynam[:5]), 2)} кОм')
dynamRi = np.mean(1/kanoddynam[:5])
dynamSigmaRi = np.var(1/kanoddynam[:5])
pd.DataFrame.from_dict(d, orient="index")
```

```python
row1 = pd.Series(ksetdynam, index=Ua)
d = {"Динам. S": row1}
print(f'Динам. <Si> = {round(np.mean(ksetdynam), 2)}±{round(np.var(ksetdynam), 2)} мА/В')
dynamSi = np.mean(ksetdynam)
dynamSigmaSi = np.var(ksetdynam)
pd.DataFrame.from_dict(d, orient="index")
```

```python
print(f"Стат. коэф. усиления Mu = {round(statSi*statRi,2)}±{round(((statSigmaSi/statSi)**2+(statSigmaRi/statRi)**2)**0.5*statSi*statRi, 2)}")
```

```python
sigmara = ((dynamSigmaRi/dynamRi)**2+(dynamSigmaRi/(dynamRi+6.2))**2)**0.5
Sg = dynamSi*dynamRi/(dynamRi+6.2)
sigmaSg = (sigmara**2+(dynamSigmaSi/dynamSi)**2)**0.5
print(f"Динам. крутизна хар-ки Sg = {round(Sg,2)}±{round((sigmara**2+(dynamSigmaSi/dynamSi)**2)**0.5*Sg, 2)} мА/B")
```
