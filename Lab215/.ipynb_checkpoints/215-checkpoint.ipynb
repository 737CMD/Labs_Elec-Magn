```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
avog = 6.02*10**23
```

```python
q = 1.6*10**(-19)
kb = 1.38*10**(-23)
def T(U):
    return q*U/2/kb
```

```python
Ud = np.array([0.8, 0.95, 1.05, 1.1, 1.15])
Id = np.array([0.55, 1.1, 1.9, 3.2,5.0])*10**(-6)/1.1
```

```python
Te = T(Ud)
Te
```

```python
s = 0.00015*np.pi*0.006
Mi = (0.02*5/6+1/6*0.004)/avog
def Bom(T, I):
    return I/0.4/q*(Mi/2/kb/T)**0.5/s
Ne = Bom(Te, Id)
Ne
```

```python
Upit = np.array([18, 20, 22, 24, 26])
Ipit = np.array([0.05, 0.05, 0.0625, 0.08, 0.09])
Ppit = Ipit*Upit
Fot = np.array([0.04375, 0.1125, 0.2, 0.325, 0.5])
Ppit
```

```python
fig, axs = plt.subplots(1,1, figsize = (7,7))
axs.set_title("Зависимость Te(P)")
axs.set_xlabel("P, Вт")
axs.set_ylabel("Te, К")
axs.minorticks_on()
axs.scatter(Ppit, Te)
axs.plot(Ppit, Te)
axs.set_xlim(0.75, 2.5)
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.6)
```

```python
fig, axs = plt.subplots(1,1, figsize = (7,7))
axs.set_title("Зависимость Ne(P)")
axs.set_xlabel("P, Вт")
axs.set_ylabel("Ne, шт/м^3")
axs.minorticks_on()
axs.scatter(Ppit, Ne)
axs.plot(Ppit, Ne)
axs.set_xlim(0.75, 2.5)
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(20))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.6)
```

```python
fig, axs = plt.subplots(1,1, figsize = (7,7))
axs.set_title("Зависимость Ifot(P)")
axs.set_xlabel("P, Вт")
axs.set_ylabel("Ifot, мкА")
axs.minorticks_on()
axs.scatter(Ppit, Fot)
axs.plot(Ppit, Fot)
axs.set_xlim(0.75, 2.5)
axs.set_ylim(0, 0.55)
axs.xaxis.set_major_locator(MultipleLocator(0.2))
axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(20))
plt.grid(which = 'minor', linestyle='--', linewidth = 0.4)
plt.grid(which = 'major', linewidth = 0.6)
```

```python
sigmat = np.array([T(0.05), T(0.025),T(0.025),T(0.025),T(0.025)])
sigmaid = np.array([0.05, 0.05, 0.05, 0.1, 0.1])*10**(-6)/1.1
sigmane = Ne*((sigmat/Te)**2+(sigmaid/Id)**2)**0.5
print(Ne)
print(sigmane)
```
